```python
for w in range(0, len(company_index)):
    df = pd.read_excel(company_index[w]+".xlsx")#讀檔
    inputs = {
    'open': df["open"],
    'high': df["high"],
    'low':  df["low"],
    'close': df["close"],
    'volume': df["volume"].astype(float)
    }
    # uses close prices (default)
    #out1 = abstract.SMA(inputs, timeperiod=5)
    #out2 = abstract.SMA(inputs, timeperiod=10)
    #out3 = abstract.SMA(inputs, timeperiod=15)
    #out4 = abstract.SMA(inputs, timeperiod=20)
    #out5 = abstract.SMA(inputs, timeperiod=60)
    #out6 = abstract.SMA(inputs, timeperiod=240)
    #out7 = abstract.MACD(inputs,fastperiod=12)
    #out8 = abstract.RSI(inputs,timeperiod=12)
    #out9 = abstract.STOCH(inputs, fastk_period=9, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)
    out10=abstract.BBANDS(inputs, timeperiod=5, nbdevup=2, nbdevdn=2, matype=0)
    out11=abstract.CDLUPSIDEGAP2CROWS(inputs)
    out12=abstract.CDLDOJISTAR(inputs)
    out13=abstract.CDLHARAMI(inputs)
    out14=abstract.WILLR(inputs, timeperiod=14)
    out15=abstract.CDLMORNINGSTAR(inputs)
    out16=abstract.AD(inputs)
    #out17=abstract.ADOSC(inputs,fastperiod=3, slowperiod=10)
    #out18=abstract.MOM(inputs, timeperiod=10)
    #out19=abstract.CDL3INSIDE(inputs)
    #out20=abstract.CDLINVERTEDHAMMER(inputs)
    a = pd.DataFrame(out10,index=["upperband","middleband","lowerband"]).T
    b = pd.DataFrame(out11,columns=["2crows"])
    c = pd.DataFrame(out12,columns=["Doji Star"])
    d = pd.DataFrame(out13,columns=["Harami Pattern"])
    e = pd.DataFrame(out14,columns=["WILLR"])
    f = pd.DataFrame(out15,columns=["Morning Star"])
    g = pd.DataFrame(out16,columns=["AD"])
    #h = pd.DataFrame(out17,columns=["A/D Oscillator"])
    #i = pd.DataFrame(out18,columns=["Momentum"])
    #j = pd.DataFrame(out19,columns=["3 Inside Up/Down"])
    #k = pd.DataFrame(out20,columns=["Inverted Hammer"])
    pd.concat([df , a , b, c , d , e , f , g], axis=1).to_excel(company_index[w]+".xlsx", \
                                              sheet_name = company_index[w], index=False)#合併進逐一的excel檔（800）
    if(w%100==0):
        print(w)
```

```python
train_set = pd.read_excel("strategy_results_train.xlsx")
```

```python
import copy
```

```python
dataset = copy.deepcopy(train_set)
```

```python
temp = []
for i in range(0, dataset.shape[0]):
# for i in range(0, 1):
    if(i%1000==0):
        print(i, datetime.datetime.now())
    trade_company = dataset.iloc[i][0].split("_")[0]
    trade_date    = dataset.iloc[i][0].split("_")[1]
    df = pd.read_excel(trade_company+".xlsx")
    trade_date_datetime = datetime.date(int(trade_date[:4]), int(trade_date[4:]), 1) - datetime.timedelta(days=1)
    lower_index = 0
    for index in range(0, df.shape[0]):
        if(df["date"].iloc[index]< trade_date_datetime):
            lower_index = index
        else:
            break
#     print(list(df.iloc[lower_index, :]))
    temp.append(list(df.iloc[lower_index, :]))
```

```python
pd.DataFrame(temp).to_excel("train_output.xlsx", index=False)
```
